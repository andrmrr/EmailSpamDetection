## Handle numbers
Create a new column with how many number tokens are there in the text
Remove all number tokens from text

```python
from nltk.tokenize import word_tokenize
df['tokens'] = df['text_combined'].apply(word_tokenize)
```

```python
# df['numbers'] = df['tokens'].apply(lambda x: len([token for token in x if token.isnumeric()]) / len(x))
df['tokens'] = df['tokens'].apply(lambda x: [token for token in x if not token.isnumeric()])
```
